```python
tokenized = open("Tokenized-2Q1.txt", "r", encoding="utf-8")
Para = tokenized.read()
Para = Para.split("\n")
remove = ['‘', '?', '.', '’', ',']

for sentence in range(len(Para)):
    Para[sentence] = Para[sentence].split(" ")

    for word in range(len(Para[sentence])):
        if "\u200c" in Para[sentence][word]:
            Para[sentence][word] = " ".join(Para[sentence][word].split("\u200c"))

        clean = ""
        for ch in Para[sentence][word]:
            if ch not in remove:
                clean += ch
        Para[sentence][word] = clean.lower()
```

### Q2) Implement Good Turing Smoothing for all the models.
The Good Turing Model has been implemented in Q12v2. We shall use Bigram due to limitations of computation

```python
Conut_List, Probability_List = NGram(2, Train, Smoothing = None)
# We use Smoothing as None here simply as assessment clearly mentions that implementaton of Good Turing Smoothing would be for Unseen n/bi-grams
```

```python
Dict_Valid_Probability_Good_Turing = {}
c= 0
for V in Validate:
    if V == ['']:
        c +=1
        print(c)
        continue
    Dict = Good_Turing_Smoothing(2, L1 = Conut_List[0], L2 = Conut_List[1], G1 = Probability_List[0], G2 = Probability_List[1], Data = [V])
    Dict_Valid_Probability_Good_Turing.update(Dict)
Dict_Valid_Probability_Good_Turing
```

```python
Dict_Test_Probability_Good_Turing = {}
c = 0
for T in Test:
    if T == ['']:
        c +=1
        print(c)
        continue
    Dict = Good_Turing_Smoothing(2, L1 = Conut_List[0], L2 = Conut_List[1], G1 = Probability_List[0], G2 = Probability_List[1], Data = [T])
    Dict_Test_Probability_Good_Turing.update(Dict)
    c +=1
    print(c)
Dict_Test_Probability_Good_Turing
```
